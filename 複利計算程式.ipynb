```python
import matplotlib.pyplot as plt

def 複利式存錢(本金,期數,利息,每期固定存入):
  total = 0
  total += 本金 # 一開始只有本金
  history = []
  for i in range(期數):
    total = total*(1+利息) # 每個月滾利息
    total += 每期固定存入 # 然後再把薪水拿一些存進去
    history.append(total)
  plt.plot(history)
  return total

def 死存錢式存錢(本金,期數,每期固定存入):
  total = 0
  total += 本金
  history = []
  for i in range(期數):
    total += 每期固定存入
    history.append(total)
  plt.plot(history)
  return total
```

# 以一年來看

```python
複利式存錢獲利 = 複利式存錢(8000,12,0.03,300) #本金8000美,12個月,每個月算3%利息就好,每個月固定加存300美
死存錢式存錢獲利 = 死存錢式存錢(8000,12,300)
plt.show()
複利式存錢獲利,死存錢式存錢獲利
```
